```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
import numpy as np
import csv
```

```python
# f* от числа шагов  |x|=8  sphere

x = np.linspace(1, 1000, 1000)

# rs = 50   nt = 5  s = 0.03
with open('../res_gso.csv', newline='') as csvfile:
    data1 = list(csv.reader(csvfile))
flat_list1 = [-float(item) for sublist in data1 for item in sublist]
x1 = np.linspace(1, len(flat_list1), len(flat_list1))
plt.plot(x1, flat_list1, '-', color="black", label="$GSO$")

# rs = 50  nt = 50    lambda = 50  eps = 0.1  s = 0.2
with open('../res_siggso.csv', newline='') as csvfile:
    data2 = list(csv.reader(csvfile))
flat_list2 = [-float(item) for sublist in data2 for item in sublist]
x2 = np.linspace(1, len(flat_list2), len(flat_list2))
plt.plot(x2, flat_list2, '--', color="black", label="$SIGGSO$")

plt.xlabel("$t$")
plt.ylabel("$\~{f^{*}}$", rotation=0)
plt.legend()
plt.grid()
plt.legend()
plt.show()
```

```python
# eps_x размерности  rastr

fig = plt.figure()
x = [2,4,8,16,32,64]
a = np.arange(len(x))

# rs = 50   nt = 5  s = 0.03
flat_list1 = [3.490015e-01, 1.245465e+00, 1.990812e+00, 2.507823e+00, 3.567206e+00, 5.294857e+00]
plt.plot(a, flat_list1, '-', marker=".", color="black", label="$GSO$")

# rs = 100  nt = 50    lambda = 1  eps = 0.1  s = 0.05
flat_list2 = [1.613057e-01, 1.132030e+00, 1.861357e+00, 2.650589e+00, 4.332961e+00, 9.941663e+00]
plt.plot(a, flat_list2, '--', marker=".", color="black", label="$SIGGSO$")

plt.xticks(a, labels=x)
plt.xlabel("$|X|$")
plt.ylabel(r"$\bar{\epsilon_{x}}$", rotation=0)
plt.legend()
plt.grid()
plt.legend()
plt.show()
```
